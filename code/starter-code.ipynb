#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
#THIS FUNCTION TAKES IN A VALUE AND RETURNS THE STANDARD DEVIATION

# LOOP THROUGH LIST, WHICH IS X OR THE COLUMN OF THE DATA FRME AND CALCULATE THE STANDARD DEVIATION
import math
def std_dev(x):
    
    std = 0
    for i in x:
        #std = math.sqrt(np.sum(abs(i - np.mean(x))**2))/(len(x)
                                                        
        std+= (i-np.mean(x))**2
        
    std= (std/ len(x)) ** (1/2)
    ##std = np.std(x)
    return std

##CALL FUNCTION WITH DF COLUMN TO CHECK AGAINST DESCRIBE
std_dev(final_df['english_2017_act'])
````

#### Plot and interpret boxplots

For each numeric variable in the dataframe create a boxplot using Seaborn. Boxplots demonstrate central tendency and spread in variables. In a certain sense, these are somewhat redundant with histograms, but you may be better able to identify clear outliers or differences in IQR, etc.

Multiple values can be plotted to a single boxplot as long as they are of the same relative scale (meaning they have similar min/max values).

Each boxplot should:
- Only include variables of a similar scale
- Have clear labels for each variable
- Have appropriate titles and labels

````python
##Box plot to show distribution of participation. seems like most of the distribution is closer to the max. we have 17 observations of 100%

c = 'green'
cc = 'red'

plt.figure(figsize= (10,10))
plt.boxplot([final_df['participation_2017_act'], final_df['participation_act_2018']], patch_artist= True,boxprops=dict(facecolor=c, color=c),
            capprops=dict(color=c),
            whiskerprops=dict(color=c),
            flierprops=dict(color=c, markeredgecolor=c),
            medianprops=dict(color=cc),
            );
    
plt.title('Distribution of ACT Participation')

plt.xlabel('YEAR')

plt.ylabel('Participation' )


plt.xticks([1, 2], ['2017', '2018'])
````
